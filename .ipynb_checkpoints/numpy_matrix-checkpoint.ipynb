```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import eqns
```

```python
df = pd.read_csv('./data/noise_parabolic_data.csv')
x_data = np.array(df.X)
y_data = np.array(df.y)
```

```python
# first, construct a coefficient matrix, A, consisting of 1, x, and x^2
A = np.ones([2000, 3])
row_num = 0
for row in A:
    row[1] = x_data[row_num]
    row[2] = x_data[row_num] ** 2
    row_num += 1
A
```

```python
# set b equal to the dependent data entries
b = np.zeros([2000, 1])
row_num = 0
for row in b:
    row[0] = y_data[row_num]
    row_num += 1
b
```

```python
# find the transpose of A matrix multiplied with A
AtA = np.dot(np.transpose(A), A)
AtA
```

```python
# find the transpose of A matrix multiplied with b
Atb = np.dot(np.transpose(A), b)
Atb
```

```python
# solve the normal equation AtA * x = Atb to give the optimized paramaters, x
x = np.linalg.solve(AtA, Atb)
x
```

```python
# generate the values of an equation using the optimized parameters
y_model = np.zeros([2000, 1])
row_num = 0
for value in y_model:
    y_model[row_num] = x[2] * x_data[row_num] ** 2 + x[1] * x_data[row_num] + x[0]
    row_num += 1
y_model
```

```python
popt = [4.99943942e-01, -2.10440025e-04, -4.63244206e+00]

def model_f(x, a, b, c):
    return a * x ** 2 + b * x + c

r_sq = eqns.r_squared(x_data, y_data, model_f, *popt)
plt.scatter(x_data, y_data, color='black')
plt.plot(x_data, y_model, color='red')
plt.title(f'R^2 = {r_sq}')
plt.show()
```
